## PREPROCESAMIENTO

Los textos deben ser transformados a un formato entendible por un algoritmo de aprendizaje autom√°tico.

Tambi√©n vamos a aplicaremos algunas tareas b√°sicas de PLN para limpiar y normalizar el texto:

- tokenizaci√≥n
- transformar a min√∫sculas
- eliminar stopwords
- aplicar stemming sobre cada token, para obtener su ra√≠z.
- eliminar cualquier palabra que contenga alg√∫n d√≠gito o signo de puntuaci√≥n.
- limpieza de emojis, enlaces, hashtags, menciones y espacios en blanco adicionales

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer

nltk.download('stopwords')
nltk.download('punkt')

stopwords_es = set(stopwords.words('spanish'))
stemmer = SnowballStemmer('spanish')

def clean_text(text):
    # Convertir a min√∫sculas
    text = str(text).lower()

    # Eliminar emojis
    text = re.sub(r'[\U00010000-\U0010ffff]', '', text)

    # Eliminar enlaces
    text = re.sub(r'http\S+', '', text)

    # Eliminar hashtags
    text = re.sub(r'#\w+', '', text)

    # Eliminar menciones
    text = re.sub(r'@\w+', '', text)

    # Eliminar espacios en blanco adicionales
    text = re.sub(r'\s+', ' ', text)

    # Tokenizar
    tokens = word_tokenize(text)

    # Eliminar stopwords
    tokens = [word for word in tokens if word not in stopwords_es]

    # Usar stems
    tokens = [stemmer.stem(word) for word in tokens]

    # Ignorar palabras que contienen d√≠gitos o s√≠mbolos especiales
    tokens = [word for word in tokens if not any(char.isdigit() or not char.isalnum() for char in word)]

    # Eliminar palabras con menos de 3 caracteres
    tokens = [word for word in tokens if len(word) >= 3]

    return tokens
```

```python
import pandas as pd
import re
import unicodedata

# Expresi√≥n regular para detectar emojis
emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticonos
                           u"\U0001F300-\U0001F5FF"  # s√≠mbolos y pictogramas
                           u"\U0001F680-\U0001F6FF"  # transporte y s√≠mbolos de mapas
                           u"\U0001F1E0-\U0001F1FF"  # banderas (iOS)
                           "]+", flags=re.UNICODE)

# Funci√≥n para limpiar un tweet
def limpiar_tweet(tweet):
    # Convertir a min√∫sculas
    tweet = tweet.lower()

    # Eliminar emojis
    tweet = re.sub(r'[\U00010000-\U0010ffff]', '', tweet)

    # Eliminar URLs
    tweet = re.sub(r'http\S+', '', tweet)

    # Eliminar hashtags
    tweet = re.sub(r'#\w+', '', tweet)

    # Eliminar menciones
    tweet = re.sub(r'@\w+', '', tweet)

    # Eliminar caracteres especiales excepto letras, espacios y emojis
    tweet = re.sub(r'[^a-zA-Z√±√ë√°√©√≠√≥√∫√Å√â√ç√ì√ö\s]+', '', tweet)

    # Reemplazar palabras con acentos por su equivalente sin tilde
    tweet = ''.join((c for c in unicodedata.normalize('NFD', tweet) if unicodedata.category(c) != 'Mn'))

    # Eliminar espacios en blanco extras
    tweet = re.sub(r'\s+', ' ', tweet)

    return tweet.strip()
```

```python
#Con b√∫squeda de HPO
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score, make_scorer



pipeline = Pipeline([
    ('bow', CountVectorizer(analyzer=clean_text)),
    ('tf', TfidfTransformer()),
    ('svm', SVC()),
])

# Definimos la cuadr√≠cula de hiperpar√°metros a explorar. A PARTIR DE VARIOS EXPERIMENTOS SE OBTUVIERON LOS HIPERPAR√ÅMETROS ESTABLECIDOS
parameters = {
    'svm__C': [10], #[0.1, 1, 10, 100],
    'svm__gamma':  [0.1] #[1, 0.1, 0.01]
    #'svm__kernel': ['rbf', 'linear', 'poly']
}

# Usamos GridSearchCV para buscar la mejor combinaci√≥n de hiperpar√°metros
# grid_search = GridSearchCV(pipeline, param_grid=parameters, cv=5, n_jobs=-1)


# Especifica la funci√≥n de puntuaci√≥n personalizada como la m√©trica para GridSearchCV
grid_search = GridSearchCV(pipeline, param_grid=parameters, cv=5, scoring='f1', n_jobs=-1)

# SE AJUSTA EL MODELO A LOS DATOS DE train1: ESTOS DATOS INCLUYEN X_trainT + Datos sint√©ticos
grid_search.fit(X_train1, y_train1)

# Mostramos los mejores hiperpar√°metros encontrados
print("Mejores hiperpar√°metros encontrados:")
print(grid_search.best_params_)
```
